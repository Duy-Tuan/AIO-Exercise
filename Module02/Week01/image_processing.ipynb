```python
import matplotlib.image as mpimg
import numpy as np
import matplotlib.pyplot as plt

def convert_gray_lightness(vector):
    return (np.max(vector) + np.min(vector)) / 2


img = mpimg.imread("dog.jpeg")
img = img.astype(np.float32)
# gray_img_01 = np.apply_along_axis(convert_gray_lightness, axis=2, arr=img) # 102.5
gray_img_01 = (np.max(img, axis=2) + np.min(img, axis=2)) // 2
gray_img_01[0, 0]
```

```python
plt.imshow(gray_img_01, cmap="gray")
```

## Question 14

```python
import matplotlib.image as mpimg
import matplotlib.pyplot as plt

def convert_gray_luminosity(vector):
    return vector[0] * 0.21 + vector[1] * 0.72 + vector[2] * 0.07


img = mpimg.imread("dog.jpeg")
img = img.astype(np.float32)


gray_img_03 = np.apply_along_axis(convert_gray_luminosity, axis=-1, arr=img)
gray_img_03[0][0]
```

```python
plt.imshow(gray_img_03, cmap="gray")
```

# Exercise 3

```python
import pandas as pd
import numpy as np

df = pd.read_csv("advertising.csv")
data = df.to_numpy()
```

```python
data.shape
```

## Question 15

```python
sale_data = data[:, 3]
max_value = np.max(sale_data)
max_index = np.argmax(sale_data)
print(f"Max: {max_value} - Index: {max_index}")
```

## Question 16

```python
tv_data = data[:, 0]
average_tv = np.mean(tv_data)
print(f"Average of TV columns: {average_tv}")
```

## Question 17

```python
sale_data = data[:, 3]

number_of_sale_gte_20 = np.sum(sale_data >= 20)
print(f"Number of sale  columns greater than or equal 20: {number_of_sale_gte_20}")
```

## Question 18

```python
sale_data = data[:, 3]
radio_data = data[:, 1]

sale_gte_15 = sale_data >= 15
average_radio_cond = np.mean(radio_data[sale_gte_15])
print(f"Average of radio data meet the condition sale >= 15: {average_radio_cond}")
```

## Question 19

```python
newspaper_data = data[:, 2]
sale_data = data[:, 3]

average_newspaper = np.mean(newspaper_data)
sum_of_sale_cond = np.sum(sale_data[newspaper_data > average_newspaper])
print(f"Number of sales columns meet the condition newspaper greater than average of newspaper: {sum_of_sale_cond}")
```

## Question 20

```python
sale_data = data[:, 3]
average_sale_data = np.mean(sale_data)
scores = np.where(sale_data > average_sale_data, "Good", "Bad")
print(scores[7:10])
```
