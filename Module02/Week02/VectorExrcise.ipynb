## 1. Math operations on matrices and vectors

### 1.1 Length of a vector

- Vector: $\mathbf{v} = [v_1, v_2, ..., v_n]^T\$
  
- Length of a vector: $\|\mathbf{v}\| = \sqrt{v_1^2 + v_2^2 + ... + v_n^2}$

```python
import numpy as np

def compute_vector_length(vector):
    len_of_vector = len(vector)
    return len_of_vector
```

```python
vector = np.arange(10)
print(f"Vector length: {compute_vector_length(vector)}")
```

### 1.2 Dot Product


- Vector: $\mathbf{v} = \begin{bmatrix} v_1 \\ v_2 \\ \vdots \\ v_n \end{bmatrix}  \quad \mathbf{u} = \begin{bmatrix} u_1 \\ u_2 \\ \vdots \\ u_n \end{bmatrix}$

- Dot Product: $\mathbf{v} \cdot \mathbf{u} = v_1 \cdot u_1 + v_2 \cdot u_2 + \ldots + v_n \cdot u_n$


```python
def compute_dot_product(vector1, vector2):
    result = np.dot(vector1, vector2)
    return result
```

```python
vector1 = np.arange(5)
vector2 = np.arange(5)

print(f"Dot product: {compute_dot_product(vector1, vector2)}")
```

### 1.3 Multiplying a vector by a matrix

- Matrix: 
  $
  \mathbf{A} = \begin{bmatrix} 
  a_{11} & \cdots & a_{1n} \\ 
  \vdots & \ddots & \vdots \\ 
  a_{m1} & \cdots & a_{mn} 
  \end{bmatrix}, \quad \mathbf{A} \in \mathbb{R}^{m \times n}
  $

- Vector: 
  $
  \mathbf{v} = \begin{bmatrix} 
  v_1 \\ 
  v_2 \\ 
  \vdots \\ 
  v_n 
  \end{bmatrix}, \quad \mathbf{v} \in \mathbb{R}^n
  $

- $\mathbf{c} = \mathbf{A} \mathbf{v} = 
  \begin{bmatrix} 
  a_{11} \cdot v_1 + \cdots + a_{1n} \cdot v_n \\ 
  \vdots \\ 
  a_{m1} \cdot v_1 + \cdots + a_{mn} \cdot v_n 
  \end{bmatrix}$

  $\mathbf{c} \in \mathbb{R}^n$

```python
def matrix_multi_vector(matrix, vector):
    result = np.dot(matrix, vector)
    return result
```

### 1.4 Multiplying a matrix by a matrix

- Matrix A: 
  $
  \mathbf{A} = \begin{bmatrix} 
  a_{11} & \cdots & a_{1n} \\ 
  \vdots & \ddots & \vdots \\ 
  a_{m1} & \cdots & a_{mn} 
  \end{bmatrix}, \quad \mathbf{A} \in \mathbb{R}^{m \times n}
  $

- Matrix B: 
  $
  \mathbf{B} = \begin{bmatrix} 
  b_{11} & \cdots & b_{1k} \\ 
  \vdots & \ddots & \vdots \\ 
  b_{n1} & \cdots & b_{nk} 
  \end{bmatrix}, \quad \mathbf{B} \in \mathbb{R}^{n \times k}
  $
- $\mathbf{C} = \mathbf{AB} = 
  \begin{bmatrix} 
  a_{11} \cdot b_{11} + \cdots + a_{1n} \cdot b_{n1} & \cdots & a_{11} \cdot b_{1k} + a_{1n} \cdot b_{nk} \\ 
  \vdots & \ddots & \vdots \\ 
  a_{m1} \cdot b_{11} + \cdots + a_{mn} \cdot b_{n1} & \cdots & a_{m1} \cdot b_{1k} + a_{mn} \cdot b_{nk} 
  \end{bmatrix}, \quad \mathbf{C} \in \mathbb{R}^{m \times k}
  $

```python
import numpy as np

def matrix_multi_matrix(matrix1, matrix2):
    result = np.dot(matrix1, matrix2)
    return result
```

### 1.5 Matrix inverse

## 1.5. Matrix inverse

- Matrix A: 
  $
  \mathbf{A} = \begin{bmatrix} 
  a & b \\ 
  c & d 
  \end{bmatrix}, \quad \mathbf{A} \in \mathbb{R}^{2 \times 2}
  $

- Determinant of **A**: 
  $
  \text{det}(\mathbf{A}) = ad - bc
  $

- If $\text{det}(\mathbf{A}) \neq 0$, **A** is invertible

- Inverse Matrix: 
  $
  \mathbf{A}^{-1} = \frac{1}{\text{det}(\mathbf{A})} \begin{bmatrix} 
  d & -b \\ 
  -c & a 
  \end{bmatrix}
  $


```python
import numpy as np

def inverse_matrix(matrix):
    result = np.linalg.inv(matrix)
    return result
```

```python
A = np.array([[-2, 6], [8, -4]])

print(inverse_matrix(A))
```

## 2. Eigenvector and eigenvalues

- $\mathbf{A} \in \mathbb{R}^{n \times n}, \mathbf{I} (identity \space matrix) \in \mathbb{R}^{n \times n}, \mathbf{v} \in \mathbb{R}^n$
- Eigenvalue $(\lambda): \text{det}(\mathbf{A} - \lambda \mathbf{I}) = 0$
- Eigenvector $(mathbf{v}): \mathbf{A} \mathbf{v} = \lambda \mathbf{v} \iff (\mathbf{A} - \lambda \mathbf{I}) \mathbf{v} = 0$
- Normalize vector: 
  $
  \frac{\mathbf{v}}{\|\mathbf{v}\|}, \quad v_i = \frac{v_i}{\sqrt{\sum_{i=1}^n v_i^2}}
  $


```python
def compute_eigenvalues_eigenvectors(matrix):
    eigenvalues, eigenvectors = np.linalg.eig(matrix)
    return eigenvalues, eigenvectors
```

```python
A = np.array([[0.9, 0.2],
              [0.1, 0.8]])

eigenvalues, eigenvectors = compute_eigenvalues_eigenvectors(A)
print(f"Eigenvalues: {eigenvalues}")
print(f"Eigenvectors: \n{eigenvectors}")
```

## 3. Cosine Similarity

- Data $(vector \space \mathbf{x}, \mathbf{y})$: 
  $
  \mathbf{x} = \{x_1, \ldots, x_N\}, \quad \mathbf{y} = \{y_1, \ldots, y_N\}
  $

- Cosine Similarity: 
  $
  \text{cs}(\mathbf{x}, \mathbf{y}) = \frac{\mathbf{x} \cdot \mathbf{y}}{\|\mathbf{x}\| \|\mathbf{y}\|} = \frac{\sum_{i=1}^n x_i y_i}{\sqrt{\sum_{i=1}^n x_i^2} \sqrt{\sum_{i=1}^n y_i^2}}
  $

```python
def compute_cosine(v1, v2):
    cos_sim = np.dot(v1, v2) / (np.linalg.norm(v1) * np.linalg.norm(v2))
    return cos_sim
```

```python
x1 = np.array([1, 2, 3, 4])
x2 = np.array([1, 0, 3, 0])

print(f"Cosine similarity: {compute_cosine(x1, x2)}")
```

## 4. Background subtraction

```python
import numpy as np
import matplotlib.pyplot as plt 
import cv2

bg1_img = cv2.imread("imgs/GreenBackground.png", 1)
bg1_img = cv2.resize(bg1_img, (678, 381))
bg1_img = bg1_img[:, :, ::-1]
plt.imshow(bg1_img)
```

```python
ob_img = cv2.imread("imgs/Object.png", 1)
ob_img = cv2.resize(ob_img, (678, 381))
ob_img = ob_img[:, :, ::-1]
plt.imshow(ob_img)
```

```python
bg2_image = cv2.imread("imgs/NewBackground.jpg", 1)
bg2_image = cv2.resize(bg2_image, (678, 381))
bg2_image = bg2_image[:, :, ::-1]
plt.imshow(bg2_image)
```

```python
def compute_difference(bg_img, input_img):
    # Convert image to gray image
    bg_img = cv2.cvtColor(bg_img, cv2.COLOR_RGB2GRAY)
    input_img = cv2.cvtColor(input_img, cv2.COLOR_RGB2GRAY)
    difference_single_channel = np.abs(input_img - bg_img)
    return difference_single_channel
```

```python
difference_single_channel = compute_difference(bg1_img, ob_img)
plt.imshow(difference_single_channel, cmap="gray")
```

```python
def compute_binary_mask(difference_single_channel):
    ret, difference_binary = cv2.threshold(difference_single_channel, 5, 255, cv2.THRESH_BINARY) 
    return difference_binary

binary_mask = compute_binary_mask(difference_single_channel)
plt.imshow(binary_mask, cmap="gray")
```

```python
def replace_background(bg1_image, bg2_image, ob_image):
    difference_single_channel = compute_difference(bg1_image , ob_image)
    binary_mask = compute_binary_mask(difference_single_channel)
    binary_mask_3dim = cv2.cvtColor(binary_mask,cv2.COLOR_GRAY2RGB)
    output = np.where(binary_mask_3dim==255, ob_image, bg2_image)
    return output
```

```python
replaced_bg_img = replace_background(bg1_img, bg2_image, ob_img)
plt.imshow(replaced_bg_img)
```
